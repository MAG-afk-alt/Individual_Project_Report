## Research Question

**Broad Question:**  
What player characteristics and behaviors are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?

**Specific Question:**  
Can player attributes such as age and total hours played predict whether a player subscribes to the newsletter?

**How the data helps:**  
- `players.csv` contains `age` and `played_hours` for each player. 
- By analyzing these two variables, we can explore whether age and hours played are significant predictors of a players likelihood to subscribe to the newsletter.


```r
#Wrangling Data

#Cleaning the 'players.csv' dataset from any na's
players_na_filtered <- players |>
    drop_na(Age, played_hours, gender)

#Wrangling the start and end time format for 'sessions.csv' dataset
sessions_dttm <- sessions|>
    mutate(start_time = dmy_hm(start_time), end_time = dmy_hm(end_time)) |>
    drop_na(start_time, end_time)
sessions_dttm

#Combining both dataset into one
combine_df <- inner_join(players_na_filtered, sessions_dttm, by = join_by(hashedEmail))
combine_df <- distinct(combine_df, .keep_all = TRUE)
combine_df

#Calculate the mean of Quantitative Variables in 'player.csv'
mean_values <- players_na_filtered |>
    summarise(mean_age = mean(Age, na.rm = "TRUE"), mean_played_hours = mean(played_hours, na.rm = "TRUE"))
mean_values

#Visualization
# Visualization 1: Histogram Count of Age
ggplot(players_na_filtered, aes(x = Age)) +
    geom_histogram(fill = "lightblue", alpha = 0.7, binwidth = 1) +
    labs(title = "Distribution of Player Age", x = "Player's Age", y = "Count") +
    theme_minimal()

# Visualization 2: Histogram Count of Played Hours 
ggplot(players_na_filtered, aes(x = played_hours)) +
    geom_histogram(fill = "blue", alpha = 0.5, binwidth = 3) +
    labs(title = "Distribution of Player's Played Hours", x = "Played Hours", y = "Count") +
    theme_minimal()

# Scatter plot with adjusted limits
ggplot(players_na_filtered, aes(x = Age, y = played_hours, color = subscribe)) +
  geom_point(alpha = 0.6) +  # scatter plot with some transparency
  labs(title = "Scatter Plot of Age vs Played Hours by Subscription Status", x = "Age of Player", y = "Total Played Hours", color = "Subscription Status") +
  theme_minimal()
```

## Insights from Visualization

**Player Count:**  
- Age wise, majority of the players are below 20 years old, and the player count goes down as age increases. This indicates that the game is more popular among younger players.
- Hours played, majority of the players play for less than 10 hours while only a few players reached more than 50 hours of gameplay. This indicates that some players are more casual while the select few are very invested in the game.

**Subscription Status:**  
The scatter plot shows a relationship between age, hours played, and subscription status. It appears that:
- Younger players tend to play more hours compared to older players, and those who play more have a higher likelihood of subscribing to the newsletter.
- For older players, although some play more hours than others in their age group, it appears that the older the players are, the less likely they are to subscribe to the newsletter.

## Methods and Plan
**Method: K-Nearest Neighbors(KNN) Classification** 
We will use KNN to predict whether a player subscribes to the newsletter or not based on `age` and `played_hours`.

**Why KNN is appropriate:**
- Simple to interpret.
- Works well with smaller datasets

**Assumptions:**
- Players with similar age and hours played behave similarly.

**Weakness:**
- Performance depends heavily on choosing the appropriate k.
- Can be affected by class imbalance

**Compare and Model selection:**
- Test multiple k-values.
- Compare each k by calculating accuracy, precision, and recall on the training data.
- Select the k-value with the best performance metrics.

**Processing the Data:**
- Impute the missing values and scale numeric variables.
- Encode subscription as a binary variable (1/0).
- Split the dataset 75% training and 25% testing.
- Perform 5 fold cross-validation on the training set to avoid data leakage.
